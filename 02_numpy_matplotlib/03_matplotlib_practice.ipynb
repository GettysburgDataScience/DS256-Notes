### Downloading the data

After finding the data URL, we use ```np.genfromtxt()``` to bring the data into numpy.

**Remember**, numpy arrays may only contain one data type! So first, we'll look at the headers to get a sense of the data each column contains. Column headers are generally in the top row of the data file.

Here is the [np.genfromtxt documentation](https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html)

````python
qb_contracts_git = 'https://raw.githubusercontent.com/GettysburgDataScience/datasets/refs/heads/main/qb_contracts.csv'
qb_contracts_headers = np.genfromtxt(qb_contracts_git, 
                                     delimiter = ',', 
                                     dtype=str, 
                                     max_rows=1)


print_list(qb_contracts_headers)
````

````python



player = np.genfromtxt(qb_contracts_git, 
                       usecols = 0, skip_header = 1, 
                       dtype=str,  delimiter = ',')

year_signed = np.genfromtxt(qb_contracts_git, 
                            usecols = 4, skip_header = 1, 
                            dtype=int, delimiter = ',')

apy = np.genfromtxt(qb_contracts_git, usecols = 7, 
                    skip_header = 1, 
                    dtype=float, delimiter = ',')
````

````python
year_mask = year_signed != 0

year_mask
````

````python
year_signed = year_signed[year_mask]
apy = apy[year_mask]
````

````python
fig, ax = plt.subplots(1,1, figsize = (7, 4))

ax.scatter(year_signed, apy,
          s = 10, c = 'coral',
          alpha = 0.5)

ax.set(
    xticks = np.arange(1980,2026, 4),
    xlabel = 'Year signed',
    ylabel = 'Contract APY ($M)',
    title = 'QB Contracts APY Over Time'
)
    
plt.show()
````

## Visualizing Gettysburg Weather Trends

````python
gettysburg_weather_git = 'https://raw.githubusercontent.com/GettysburgDataScience/datasets/refs/heads/main/gettysburg_weather.csv'
gettysburg_weather_headers = np.genfromtxt(gettysburg_weather_git, dtype=str, max_rows=1, delimiter = ',')

print_list(gettysburg_weather_headers)
````

````python
date = np.genfromtxt(gettysburg_weather_git, 
                     usecols = 1, skip_header=1,
                     dtype=np.datetime64,
                     delimiter = ',')

tmax = np.genfromtxt(gettysburg_weather_git, 
                     usecols = 4, skip_header=1,
                     dtype=int, missing_values = np.nan, 
                     delimiter = ',')

tmin = np.genfromtxt(gettysburg_weather_git, 
                     usecols = 5, skip_header=1,
                     dtype=int, missing_values = np.nan,
                     delimiter = ',')
````

## Visualizing Gun Deaths in the United States

````python
gun_deaths_git = 'https://raw.githubusercontent.com/GettysburgDataScience/datasets/refs/heads/main/gun_deaths_cdc.csv'
gun_deaths_headers = np.loadtxt(gun_deaths_git, dtype=str, max_rows=1, delimiter = ',')

print_list(gun_deaths_headers)
````

````python


state = np.genfromtxt(gun_deaths_git, delimiter = ',',
                       dtype = str, 
                      usecols = 0, skip_header = 1)

year = np.genfromtxt(gun_deaths_git, delimiter = ',',
                       dtype = int, 
                      usecols = 2, skip_header = 1)

deaths = np.genfromtxt(gun_deaths_git, delimiter = ',',
                       dtype = int, 
                      usecols = 3, skip_header = 1)

popu = np.genfromtxt(gun_deaths_git, delimiter = ',',
                       dtype = int, 
                      usecols = 4, skip_header = 1)

````

````python
popu_100k = popu/100000
````

````python
death_per_100k = deaths/popu_100k
````

````python
year_mask = year==2024

popu_100k = popu_100k[year_mask]
death_per_100k = death_per_100k[year_mask]
state = state[year_mask]
````

````python

fig, ax = plt.subplots(1,1, figsize = (7,7))

ax.barh(np.arange(len(state)), death_per_100k)

ax.set(
    yticks = np.arange(len(state)),
    yticklabels = state)
plt.show()
````
