## Rules of Wordle

If you haven't played Wordle before, try it [here](https://www.nytimes.com/games/wordle/index.html). Players 

Players get 6 chances to guess a 5-letter secret word. Each guess must be a 5-letter word. Players get feedback as to which letters in the word are correct:

 - GREEN letter - The letter is correct and in the correct position
 - YELLOW letter - The letter is in the word, but in the incorrect position
 - GRAY letter - The letter is not in the word at all

The game ends when either the player has guessed the secret word or the player has made 6 incorrect guesses.

### Some useful tools

*USEFUL TOOL:* ```zip(list_a, list_b, list_c)``` can be used in a for loop to iterate over several lists in the same loop. In the first iteration, it will capture the first elements of each list. In the second iteration, it will capture the second element of each list. And so on.

```python
for a, b in zip('hello', 'there'):
    print(a+b)
```

would capture the letters of 'hello' in the variable ```a``` and the letters of 'there' in the variable ```b``` to print:

```
ht
eh
le
lr
oe
```


*USEFUL TOOL:* ```enumerate(list_a)``` returns a counter variable along with the items of list_a.

```python
for k, a in enumerate ('hello'):  # k will be the counter variable
    print(f'{k}. {a}')
```

would capture the letters of 'hello' in the variable ```a``` and the letters of 'there' in the variable ```b``` to print:

```
1. h
2. e
3. l
4. l
5. o
```
