**1. Academic Standing Notifications**


Write the function ```student_advisory``` that takes as input a student’s grade and using a set of if statements, prints a message for the student.
 - If a student is earning above an 80, they are in good standing and nothing needs to be done.
 - If they are earning between 60-80, they're struggling and should get a note to attend office hours.
 - If they are below 60, they are failing and should get a notification to schedule a one-on-one with the professor.

````python
def student_advisory(grade):
    if grade >= 80:
        print("You're doing well in the course!")
    elif grade >= 60:
        print('You should attend office hours.')
    else:
        print('Please schedule a meeting with your instructor.')
````

**2. Leap Year?**
Write the function is_leap_year that takes as input a year and returns whether or not it’s a leap year. Here are the rules for leap years.
 - Divisible by 4 but not divisible by 100.
 - OR divisible by 400.

````python
def is_leap_year(year):
    if year % 400 == 0 or (year % 4 == 0 and not year % 100 == 0):
        return True
    else:
        return False
````
